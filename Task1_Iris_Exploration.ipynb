# Task 1: Exploring and Visualizing the Iris Dataset

## Introduction
The objective of this task is to explore and visualize the Iris dataset using Python.
The dataset contains measurements of iris flowers and is commonly used for
understanding basic data analysis and visualization techniques.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title('Sepal Length vs Petal Length')
plt.show()
```

```python
df['sepal_length'].hist(bins=20)
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')
plt.show()
```

### Conclusion
From the visualizations, we observe that petal length and petal width clearly
differentiate between the three iris species. This indicates that these features
are highly important for classification.



```python
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Petal Length by Species')
plt.show()
```
